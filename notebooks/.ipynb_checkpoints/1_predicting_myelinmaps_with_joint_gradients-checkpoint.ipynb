```python
surfaces = dict()
surfaces['human_lh'] = "/mnt/WD10TB/datasets/data_view_crossspecies/human/surfaces/S1200.L.very_inflated_MSMAll.10k_fs_LR.surf.gii"
surfaces['human_rh'] = "/mnt/WD10TB/datasets/data_view_crossspecies/human/surfaces/S1200.R.very_inflated_MSMAll.10k_fs_LR.surf.gii"
surfaces['marmoset_lh'] = "/mnt/WD10TB/datasets/data_view_crossspecies/marmoset/surfaces/surfFS.lh.pial.inflated.10k.surf.gii"
surfaces['marmoset_rh'] = "/mnt/WD10TB/datasets/data_view_crossspecies/marmoset/surfaces/surfFS.rh.pial.inflated.10k.surf.gii"

sing_shell="singularity exec -B /mnt/WD10TB -B / /home/geoff/Desktop/containers/fmriprep_ciftify-1.3.2-2.3.3"
marmoset_dataset='NIH'; THR_=0.01
jointgrad_dir = jointgrad_dir = f"../data/data_xspecies/JOINT0_SPARSE_v3/joint_eigenmap.human-100UR.marmoset-{marmoset_dataset}.thr-corr-{THR_}.norm"

def get_joint_gradient(species,gradient_id,hemi,jointgrad_dir=jointgrad_dir,sing_shell=sing_shell):
    
    def read_dscalar(dscalar,hemi):
        # Set-up tmpdir
        tmpdir = tempfile.TemporaryDirectory()
        # Read data
        left_func = f"{os.path.join(tmpdir.name,'left_func.func.gii')}"
        right_func = f"{os.path.join(tmpdir.name,'right_func.func.gii')}"
        cmd = f"{sing_shell} wb_command -cifti-separate {dscalar} COLUMN -metric CORTEX_LEFT {left_func} -metric CORTEX_RIGHT {right_func}"
        os.system(cmd)
        if hemi == 'lh':
            read_func = left_func
        elif hemi == 'rh':
            read_func = right_func
        else:
            NotImplemented
        
        return nib.load(read_func).agg_data()
        
    gradient_path = os.path.join(jointgrad_dir,f"joint_eigenmap_{species}.dscalar.nii")
    data = read_dscalar(gradient_path,hemi)
    data = data[gradient_id]
    
    return data

def view_joint_gradient(gradient_id,title=''):
    tmpdir = tempfile.TemporaryDirectory()
    
    imgs_comb = {}
    for species in ['human','marmoset']:
        """
        Create images of lh/rh & medial/lateral of `gradient_id`
        """
        fig_list = []
        # Left hemisphere
        hemi='lh'
        # Lateral view
        view_='lateral'
        gradient = get_joint_gradient(species,gradient_id,hemi)
        plotting.plot_surf(surfaces[f"{species}_{hemi}"],gradient,
                           view=view_,hemi='left',
                           cmap='jet',symmetric_cmap=False,vmin=0,vmax=1,threshold=0,
                          )
        plt.savefig(os.path.join(tmpdir.name,f"{view_}_{hemi}.jpg"),dpi=1080)
        fig_list.append(os.path.join(tmpdir.name,f"{view_}_{hemi}.jpg"))
        plt.close()
        # Medial view
        view_='medial'
        plotting.plot_surf(surfaces[f"{species}_{hemi}"],gradient,
                           view=view_,hemi='left',
                           cmap='jet',symmetric_cmap=False,vmin=0,vmax=1,threshold=0,
                          )
        plt.savefig(os.path.join(tmpdir.name,f"{view_}_{hemi}.jpg"),dpi=1080)
        fig_list.append(os.path.join(tmpdir.name,f"{view_}_{hemi}.jpg"))
        plt.close()
        # Right hemisphere
        hemi='rh'
        # Medial view
        view_='medial'
        plotting.plot_surf(surfaces[f"{species}_{hemi}"],gradient,
                           view=view_,hemi='right',
                           cmap='jet',symmetric_cmap=False,vmin=0,vmax=1,threshold=0,
                          )
        plt.savefig(os.path.join(tmpdir.name,f"{view_}_{hemi}.jpg"),dpi=1080)
        fig_list.append(os.path.join(tmpdir.name,f"{view_}_{hemi}.jpg"))
        plt.close()
        # Lateral view
        view_='lateral'
        gradient = get_joint_gradient(species,gradient_id,hemi)
        plotting.plot_surf(surfaces[f"{species}_{hemi}"],gradient,
                           view=view_,hemi='right',
                           cmap='jet',symmetric_cmap=False,vmin=0,vmax=1,threshold=0,
                          )
        plt.savefig(os.path.join(tmpdir.name,f"{view_}_{hemi}.jpg"),dpi=1080)
        fig_list.append(os.path.join(tmpdir.name,f"{view_}_{hemi}.jpg"))
        plt.close()
    
        """
        hstack lh/rh & medial/lateral views
        """
        import PIL
        from PIL import Image
        imgs = [ PIL.Image.open(i) for i in fig_list ]
        imgs_comb[species] = np.hstack([np.asarray(i) for i in imgs])
    
    fig = plt.figure(figsize=(16,8))
    plt.suptitle(f"{title}")
    fig.add_subplot(2,1,1)
    plt.imshow(imgs_comb['human'])
    plt.axis('off')
    
    fig.add_subplot(2,1,2)
    plt.imshow(imgs_comb['marmoset'])
    plt.axis('off')

for gradient_id in tqdm(range(13)):
    view_joint_gradient(gradient_id,f"Gradient-{gradient_id}")
```

# Functions

```python
def get_gradient_data(species='human',n_grads=12):
    data = {}
    for hemi in ['lh','rh']:
        data[hemi] = np.zeros((10242,n_grads))
        for gradient_id in range(1,n_grads+1):
            data[hemi][:,(gradient_id-1)] = get_joint_gradient(species,gradient_id,hemi)
            
    return data

def get_dscalar(dscalar,sing_shell=sing_shell):
    
    def read_dscalar(dscalar,hemi):
        # Set-up tmpdir
        tmpdir = tempfile.TemporaryDirectory()
        # Read data
        left_func = f"{os.path.join(tmpdir.name,'left_func.func.gii')}"
        right_func = f"{os.path.join(tmpdir.name,'right_func.func.gii')}"
        cmd = f"{sing_shell} wb_command -cifti-separate {dscalar} COLUMN -metric CORTEX_LEFT {left_func} -metric CORTEX_RIGHT {right_func}"
        os.system(cmd)
        if hemi == 'lh':
            read_func = left_func
        elif hemi == 'rh':
            read_func = right_func
        else:
            NotImplemented
        
        return nib.load(read_func).agg_data()
        
    data_lh = read_dscalar(dscalar,'lh')
    data_rh = read_dscalar(dscalar,'rh')
    
    return data_lh, data_rh  

def get_myelin_data(species='human'):
    if species == 'human':
        lh = "../data/myelinmap/S1200.lh.MyelinMap_BC_MSMAll.10k_fs_LR.s6.shape.gii"
        rh = "../data/myelinmap/S1200.rh.MyelinMap_BC_MSMAll.10k_fs_LR.s6.shape.gii"
    if species == 'marmoset':
        lh = "../data/myelinmap/surfFS.lh.myelinmap.10k.s2.shape.gii"
        rh = "../data/myelinmap/surfFS.rh.myelinmap.10k.s2.shape.gii"
    
    myelin_data = {}
    myelin_data['lh'] = nib.load(lh).agg_data()
    myelin_data['rh'] = nib.load(rh).agg_data()
    
    return myelin_data

def get_cortex_coords(species='human'):
    """
    Create cortex mask that removes outliers (outliers defined by |Z|>2.5), and medial wall (myelinmap=0)
    """
    myelin_data = get_myelin_data(species=species)

    lh_coords = np.where(myelin_data['lh'] > 0)
    rh_coords = np.where(myelin_data['rh'] > 0)
    myelin_subset_data_lh = myelin_data['lh'][lh_coords] 
    myelin_subset_data_rh = myelin_data['rh'][rh_coords]
    myelin_concat = np.concatenate((myelin_subset_data_lh,
                                    myelin_subset_data_rh))
    mean = myelin_concat.mean()
    std = myelin_concat.std()
    myelin_concat = ( myelin_concat-mean ) / std
    outlier_idx = np.where(np.abs(myelin_concat)>3.2)
    myelin_concat[outlier_idx] = -100
    
    X = np.zeros((20484,))
    for idx,coord in enumerate(np.concatenate((lh_coords[0],rh_coords[0]+10242))):
        if myelin_concat[idx] == -100:
            X[coord] = 0
        else:
            X[coord] = 1
    
    myelin_data['lh'] = X[:10242]
    myelin_data['rh'] = X[10242:]
    mask = {}
    for hemi,myelinmap in myelin_data.items(): 
        mask[hemi] = np.where(myelinmap==1)
    
    return mask     

def remap_array_onto_cortex(array,species='human'):
    
    cortex_coords = {}
    cortex_coords[species] = get_cortex_coords(species=species)
    
    X_ = np.zeros((20484,))
    for idx,coord in enumerate(np.concatenate((cortex_coords[species]['lh'][0],cortex_coords[species]['rh'][0]+10242))):
        X_[coord] = array[idx]
        
    return X_

def remap_array_onto_hemi(array,hemi,species='human'):
    print(f"{hemi}: {array.shape}")
    
    cortex_coords = {}
    cortex_coords[species] = get_cortex_coords(species=species)
    
    X_ = np.zeros((10242,))
    for idx,coord in enumerate(cortex_coords[species][hemi][0]):
        X_[coord] = array[idx]
        
    return X_

def get_prediction_data():
    """
    Training data = human myelin map (Y) / gradient (X)
    Test data = marmoset myelin map (Y) / gradient (X)
    """
    data_gradients = {}
    cortex_coords = {}
    data_myelin = {}
    for species in ['human','marmoset']:
        data_gradients[species] = get_gradient_data(species=species)
        cortex_coords[species] = get_cortex_coords(species=species)
        data_myelin[species] = get_myelin_data(species=species)
    
    Y_test = np.concatenate((data_myelin['human']['lh'][cortex_coords['human']['lh'][0]],
                              data_myelin['human']['rh'][cortex_coords['human']['rh'][0]]))
    X_test = np.vstack((data_gradients['human']['lh'][cortex_coords['human']['lh'][0],:],
                         data_gradients['human']['rh'][cortex_coords['human']['rh'][0],:]))
    X_train = np.vstack((data_gradients['marmoset']['lh'][cortex_coords['marmoset']['lh'][0],:],
                        data_gradients['marmoset']['rh'][cortex_coords['marmoset']['rh'][0],:]))
    Y_train = np.concatenate((data_myelin['marmoset']['lh'][cortex_coords['marmoset']['lh'][0]],
                         data_myelin['marmoset']['rh'][cortex_coords['marmoset']['rh'][0]]))
    Y_test = np.concatenate((data_myelin['human']['lh'][cortex_coords['human']['lh'][0]],
                              data_myelin['human']['rh'][cortex_coords['human']['rh'][0]]))
    # Z-normalize
    Y_train = (Y_train-Y_train.mean()) / Y_train.std()
    Y_test = (Y_test-Y_test.mean()) / Y_test.std()
    
    return X_train,X_test,Y_train,Y_test

def get_null_myelinmap_surrogates(species,hemi,n_surr=1000):
    """
    RUN "generate surrogate map" notebook first
    In line with multiple regression analyses, myelinmap are also 
    Z-normalized before generating surrogates.
    """
    from brainsmash.mapgen.sampled import Sampled
    from brainsmash.mapgen.memmap import txt2memmap
    # Get paths
    if species == 'marmoset' and hemi == 'lh':
        brain_map_file = "../data/myelinmap/surfFS.lh.myelinmap.10k.s2.shape.gii"
        cortex_coords = get_cortex_coords(species=species)['lh']
    if species == 'marmoset' and hemi == 'rh':
        brain_map_file = "../data/myelinmap/surfFS.rh.myelinmap.10k.s2.shape.gii"
        cortex_coords = get_cortex_coords(species=species)['rh']
    if species == 'human' and hemi == 'lh':
        brain_map_file = "../data/myelinmap/S1200.lh.MyelinMap_BC_MSMAll.10k_fs_LR.s6.shape.gii"
        cortex_coords = get_cortex_coords(species=species)['lh']
    if species == 'human' and hemi == 'rh':
        brain_map_file = "../data/myelinmap/S1200.rh.MyelinMap_BC_MSMAll.10k_fs_LR.s6.shape.gii"
        cortex_coords = get_cortex_coords(species=species)['rh']
    
    myelinmap = nib.load(brain_map_file).agg_data() # Load myelinmap
    # Get medial wall mask
    cortex_mask = np.ones((10242,))
    cortex_mask[cortex_coords] = 0
    brain_map_file = myelinmap[cortex_coords[0]] # Get rid of medial wall
    myelinmap = (brain_map_file-brain_map_file.mean()) / brain_map_file.std() #Z-normalize
    # Load distmap into memory
    dist_mat_fin = f"../data/brainsmash/{species}_{hemi}_dense_geodesic_distmap.txt"
    output_dir = "../data/brainsmash"
    output_files = txt2memmap(dist_mat_fin, '../data/brainsmash',maskfile=cortex_mask)
    # Generate surrogate myelin maps
    dist_mat_mmap = np.load(output_files['distmat'])
    index_mmap = np.load(output_files['index'])
    sampled = Sampled(brain_map_file,
                      dist_mat_mmap,
                      index_mmap)
    # Run
    surrogates = sampled(n=n_surr)
    
    return myelinmap, surrogates
```

Save actual and predicted myelin maps as giftis

```python
def save_myelin_data(actual,pred,species='human'):
    
    if species == 'human':
        lh = "../data/myelinmap/S1200.lh.MyelinMap_BC_MSMAll.10k_fs_LR.s6.shape.gii"
        rh = "../data/myelinmap/S1200.rh.MyelinMap_BC_MSMAll.10k_fs_LR.s6.shape.gii"
    if species == 'marmoset':
        lh = "../data/myelinmap/surfFS.lh.myelinmap.10k.s2.shape.gii"
        rh = "../data/myelinmap/surfFS.rh.myelinmap.10k.s2.shape.gii"
    
    for myelinmap in [lh,rh]:
        if 'lh' in myelinmap:
            hemi = 'lh'
        else:
            hemi = 'rh'
        for ix,arr in enumerate([actual,pred]):
            x = nib.load(myelinmap)
            #arr += 3
            if hemi == 'lh':
                y = remap_array_onto_cortex(arr,species=species)[:10242]
            else:
                y = remap_array_onto_cortex(arr,species=species)[10242:]
            x.add_gifti_data_array(nib.gifti.GiftiDataArray(y,datatype='NIFTI_TYPE_FLOAT32', intent='NIFTI_INTENT_POINTSET'))
            x.remove_gifti_data_array(0)
            
            if ix == 0:
                nib.save(x,f"./figures/actual_{hemi}_{species}_myelinmap.shape.gii")
            else:
                nib.save(x,f"./figures/pred_{hemi}_{species}_myelinmap.shape.gii")
                
save_myelin_data(Y_train,Y_marmoset_pred,species='marmoset')
save_myelin_data(Y_test,Y_human_pred,species='human')
```

# STATS: Perform test to see if predicted and actual myelinmaps are correlated

```python
from scipy.stats import pearsonr

cortex_coords = {}
for species in ['human','marmoset']:
    cortex_coords[species] = get_cortex_coords(species=species)

null = {}
test = {}
for species in ['marmoset','human']:
    
    for hemi in ['lh','rh']:
        
        if species == 'marmoset' and hemi == 'lh':
            pred_myelinmap = remap_array_onto_cortex(Y_marmoset_pred,species='marmoset')[:10242][cortex_coords[species][hemi]]
        if species == 'marmoset' and hemi == 'rh':
            pred_myelinmap = remap_array_onto_cortex(Y_marmoset_pred,species='marmoset')[10242:][cortex_coords[species][hemi]]
        if species == 'human' and hemi == 'lh':
            pred_myelinmap = remap_array_onto_cortex(Y_human_pred,species='human')[:10242][cortex_coords[species][hemi]]
        if species == 'human' and hemi == 'rh':
            pred_myelinmap = remap_array_onto_cortex(Y_human_pred,species='human')[10242:][cortex_coords[species][hemi]]

        null[f"{species}_{hemi}"] = np.zeros((n_surr,))
        surr = surrogates[f"{species}_{hemi}"]
        actual_myelinmap = myelinmap[f"{species}_{hemi}"] #[cortex_coords[species][hemi]]

        test[f"{species}_{hemi}"] = pearsonr(actual_myelinmap,pred_myelinmap)[0]
        for ix in range(n_surr):
            null[f"{species}_{hemi}"][ix] = pearsonr(surr[ix,:],pred_myelinmap)[0]

        pval = (test[f"{species}_{hemi}"]<null[f"{species}_{hemi}"]).sum()/n_surr
        print(f"{species} {hemi}\np-value: {pval}\n")
```

```python
fig, (ax1, ax2) = plt.subplots(ncols=2, sharey=True,sharex=True,
                               figsize=(15,4),dpi=200)

# Plot1
sns.kdeplot(null[f"marmoset_lh"],fill=True,color='k',lw=0,
            bw=0.4,ax=ax1)
sns.kdeplot(null[f"marmoset_rh"],fill=False,color='k',lw=3,
            bw=0.4,ax=ax1)

ax1.axvline(x=test[f"marmoset_lh"],c='red',lw=4)
ax1.axvline(x=test[f"marmoset_rh"],c='orange',lw=4,linestyle='-')

# Plot2
sns.kdeplot(null[f"human_lh"],fill=True,color='k',lw=0,
            bw=0.4,ax=ax2)
sns.kdeplot(null[f"human_rh"],fill=False,color='k',lw=3,
            bw=0.4,ax=ax2)
ax2.axvline(x=test[f"human_lh"],c='red',lw=4)
ax2.axvline(x=test[f"human_rh"],c='orange',lw=4,linestyle='-')

# Plotting params
for ax in [ax1,ax2]:
    ax.set_xticks([-.8,-.4,0,.4,.8])
    ax.set_xticklabels(["-0.8",'','0','','0.8'])
    ax.spines['top'].set_visible(False)
    ax.spines['right'].set_visible(False)
    ax.xaxis.set_ticks_position('bottom')
    ax.yaxis.set_ticks_position('left')
    ax.set_xlim([-.8,.8])
    ax.set_xlabel(f"Pearson correlation")
    ax.set_ylabel(f"Density",rotation='horizontal')
    ax.yaxis.set_label_coords(-.25,.45)
    ax.tick_params(length=10,width=2)
    for axis in ['top','bottom','left','right']:
        ax.spines[axis].set_linewidth(2.)
```
