```python
import os, sys
import tempfile
import nibabel as nib
import numpy as np
# Plotting
import matplotlib.pyplot as plt
from matplotlib import cm
from nilearn import plotting
import seaborn as sns
from tqdm import tqdm
# Data aggregation tools
from utils.joint_embedding import joint_gradient_landmarks

sing_shell="singularity exec -B /mnt/WD10TB -B / /home/geoff/Desktop/containers/fmriprep_ciftify-1.3.2-2.3.3"
```

```python
# Plotting function
def view_10k_array(fn_prefix,surf,surf_arg,
                   array,array_arg,
                   cmap,vmin,vmax,symmetric_cmap,
                   show_flag=False,
                   threshold=None):
    
    tmpdir = tempfile.TemporaryDirectory()
    
    """
    Create images of lh/rh & medial/lateral/dorsal of `gradient_id`
    """
    fig_list = []
    for hemi,hemi_,view_ in [('lh','left','lateral'),
                             ('lh','left','medial'),
                             ('lh','left','dorsal'),
                             ('rh','right','dorsal'),
                             ('rh','right','medial'),
                             ('rh','right','lateral')]:
        plotting.plot_surf(surf[f"{surf_arg}_{hemi}"],array[f"{array_arg}_{hemi}"], cmap=cmap,
                           view=view_,hemi=hemi_,vmin=vmin,vmax=vmax,symmetric_cmap=symmetric_cmap,
                           threshold=None)
        plt.savefig(os.path.join(tmpdir.name,f"{view_}_{hemi}.jpg"),dpi=1080)
        fig_list.append(os.path.join(tmpdir.name,f"{view_}_{hemi}.jpg"))
        plt.close()

    """
    hstack lh/rh & medial/lateral views
    """
    import PIL
    from PIL import Image
    imgs = []
    for i in fig_list:
        if 'dorsal' in i:
            imgs.append(PIL.Image.open(i).rotate(90))
        else:
            imgs.append(PIL.Image.open(i))
        
    imgs_comb = np.hstack([np.asarray(i) for i in imgs])
    
    fig = plt.figure(figsize=(16,4))
    fig.add_subplot(1,1,1)
    plt.imshow(imgs_comb)
    plt.title(f"{fn_prefix} {array_arg}")
    plt.axis('off')
    
    # SAVE FIGURE
    plt.savefig(os.path.join('figures',f"{fn_prefix}_{surf_arg}_{array_arg}.jpg"),dpi=1080)
    
    # don't plot
    if not show_flag:
        plt.close()

def get_dscalar(dscalar,sing_shell=sing_shell):
    
    def read_dscalar(dscalar,hemi):
        # Set-up tmpdir
        tmpdir = tempfile.TemporaryDirectory()
        # Read data
        left_func = f"{os.path.join(tmpdir.name,'left_func.func.gii')}"
        right_func = f"{os.path.join(tmpdir.name,'right_func.func.gii')}"
        cmd = f"{sing_shell} wb_command -cifti-separate {dscalar} COLUMN -metric CORTEX_LEFT {left_func} -metric CORTEX_RIGHT {right_func}"
        os.system(cmd)
        if hemi == 'lh':
            read_func = left_func
        elif hemi == 'rh':
            read_func = right_func
        else:
            NotImplemented
        
        return nib.load(read_func).agg_data()
        
    data_lh = read_dscalar(dscalar,'lh')
    data_rh = read_dscalar(dscalar,'rh')
    
    return data_lh, data_rh

def get_cortex_coords(cortex_mask = "../data/data_marmoset/surfFS.atlasroi.10k.dscalar.nii"):
    lh_mask, rh_mask = get_dscalar(cortex_mask)
    mask = dict()
    mask['lh'] = np.where(lh_mask == 1)
    mask['rh'] = np.where(rh_mask == 1)
    return mask        

surfaces = dict()
surfaces['human_lh'] = "/mnt/WD10TB/datasets/data_view_crossspecies/human/surfaces/S1200.L.very_inflated_MSMAll.10k_fs_LR.surf.gii"
surfaces['human_rh'] = "/mnt/WD10TB/datasets/data_view_crossspecies/human/surfaces/S1200.R.very_inflated_MSMAll.10k_fs_LR.surf.gii"
surfaces['marmoset_lh'] = "/mnt/WD10TB/datasets/data_view_crossspecies/marmoset/surfaces/surfFS.lh.pial.inflated.10k.surf.gii"
surfaces['marmoset_rh'] = "/mnt/WD10TB/datasets/data_view_crossspecies/marmoset/surfaces/surfFS.rh.pial.inflated.10k.surf.gii"

mask = get_cortex_coords()

joint_marmoset_grads = {}
for cohort in ['NIH','UWO']:
    joint_marmoset_grads[cohort] = f"../data/data_xspecies/joint_eigenmap.human-100UR.marmoset-{cohort}.thr-corr-0.001.norm/joint_eigenmap_marmoset.dscalar.nii"
    joint_marmoset_grads[f"{cohort}_lh"], joint_marmoset_grads[f"{cohort}_rh"] = get_dscalar(joint_marmoset_grads[cohort])
```

```python
"""
INPUTS
"""
joint_idx = 2
cohort = "NIH"
singlesubjdir = "../data/data_marmoset/single_subject_gradients/"
single_marmoset_dscalar = [os.path.join(singlesubjdir,i) for i in os.listdir(singlesubjdir) if 'dscalar.nii' in i ]


"""
Get joint gradient of interest: specify idx (NOTE: idx=0 is the fiedler vector)
"""
selected_joint_marmoset_grad = np.concatenate((joint_marmoset_grads[f"{cohort}_lh"][joint_idx][mask['lh']],
                                               joint_marmoset_grads[f"{cohort}_rh"][joint_idx][mask['rh']]))
"""
Plot joint gradient
"""
plot_selected_joint_marmoset_grad = {}
lh,rh = get_dscalar(joint_marmoset_grads[cohort])
plot_selected_joint_marmoset_grad['marmoset_lh'] = lh[joint_idx]
plot_selected_joint_marmoset_grad['marmoset_rh'] = rh[joint_idx]
cmap_ = cm.get_cmap('jet', 10)
view_10k_array(f"joint_grad_{cohort}_{joint_idx}",surfaces,'marmoset',plot_selected_joint_marmoset_grad,
           'marmoset',cmap_,0,1,False,show_flag=True)

""" 
Loop through each marmoset
- Get ind_idx that has the highest r: save_idx
- Save save_idx and sign_flag; sign_flag specifies +/-
"""
matched_list = []
selected_ind_marmoset_grad = {}
counter = 0
for marmoset in tqdm(single_marmoset_dscalar):
    # Loop through each gradient (only browse top 10)
    marm_id = marmoset.split('/')[-1].split('.cortex')[0]
    save_r = 0
    for ind_idx in range(1,11):
        data_lh, data_rh = get_dscalar(marmoset)
        ind_marmoset_grad = np.concatenate((data_lh[ind_idx][mask['lh']],
                                            data_rh[ind_idx][mask['rh']]))
        r = np.corrcoef(selected_joint_marmoset_grad,ind_marmoset_grad)[0,1]
        # Update save_idx
        if np.abs(r) > save_r:
            save_r = np.abs(r)
            save_idx = ind_idx
            save_flag = '+' if r > 0 else '-'
    matched_list.append((marm_id,save_idx,save_flag,save_r))
    
    """
    PLOT
    """
    lh,rh = get_dscalar(marmoset)
    if save_flag == '-':
        selected_ind_marmoset_grad['marmoset_lh'] = (lh[save_idx]*-1)+1
        selected_ind_marmoset_grad['marmoset_rh'] = (rh[save_idx]*-1)+1
    else:
        selected_ind_marmoset_grad['marmoset_lh'] = lh[save_idx]
        selected_ind_marmoset_grad['marmoset_rh'] = rh[save_idx]
    print(f"SAVE PARAMS: {marm_id}, {save_idx}, {save_flag}, {save_r}")
    view_10k_array(f"{marm_id}_{save_idx}",surfaces,'marmoset',selected_ind_marmoset_grad,
                   'marmoset',cmap_,0,1,False,show_flag=True)
```
