```python
def get_joint_gradient(species,gradient_id,hemi,jointgrad_dir=jointgrad_dir,sing_shell=sing_shell):
    
    def read_dscalar(dscalar,hemi):
        # Set-up tmpdir
        tmpdir = tempfile.TemporaryDirectory()
        # Read data
        left_func = f"{os.path.join(tmpdir.name,'left_func.func.gii')}"
        right_func = f"{os.path.join(tmpdir.name,'right_func.func.gii')}"
        cmd = f"{sing_shell} wb_command -cifti-separate {dscalar} COLUMN -metric CORTEX_LEFT {left_func} -metric CORTEX_RIGHT {right_func}"
        os.system(cmd)
        if hemi == 'lh':
            read_func = left_func
        elif hemi == 'rh':
            read_func = right_func
        else:
            NotImplemented
        
        return nib.load(read_func).agg_data()
        
    gradient_path = os.path.join(jointgrad_dir,f"joint_eigenmap_{species}.dscalar.nii")
    data = read_dscalar(gradient_path,hemi)
    data = data[gradient_id]
    
    return data

def get_dscalar(dscalar,sing_shell=sing_shell):
    
    def read_dscalar(dscalar,hemi):
        # Set-up tmpdir
        tmpdir = tempfile.TemporaryDirectory()
        # Read data
        left_func = f"{os.path.join(tmpdir.name,'left_func.func.gii')}"
        right_func = f"{os.path.join(tmpdir.name,'right_func.func.gii')}"
        cmd = f"{sing_shell} wb_command -cifti-separate {dscalar} COLUMN -metric CORTEX_LEFT {left_func} -metric CORTEX_RIGHT {right_func}"
        os.system(cmd)
        if hemi == 'lh':
            read_func = left_func
        elif hemi == 'rh':
            read_func = right_func
        else:
            NotImplemented
        
        return nib.load(read_func).agg_data()
        
    data_lh = read_dscalar(dscalar,'lh')
    data_rh = read_dscalar(dscalar,'rh')
    
    return data_lh, data_rh


def get_gradient_distribution(species,option,datadir="notebook_data/2_DMN_nodes"):
    """
    Function to grab gradient values from the somatomotor-DMN gradient
    """
    def get_roi(species,rois):
        """
        * rois is '+'-delimited
        Human ROIs: dlPFC, PCC, PPC
        Marmoset ROIs: dlPFC [A8aD,A6DR], PCC [PGM,A23a,A23b,A29d,A30],
                       PPC [LIP,MIP,AIP,VIP], mPFC [A32,A10,A32V]
        """
        if species == 'human':
            if isinstance(rois, str) and '+' not in rois:
                roi = rois
                human_dscalar = f"{datadir}/ROIs/{species}/human.HCP_Parcellation.10k.roi-{roi}.dscalar.nii"
                values = np.concatenate((get_joint_gradient(species,1,'lh')[np.where(get_dscalar(human_dscalar)[0]==1)],
                                         get_joint_gradient(species,1,'rh')[np.where(get_dscalar(human_dscalar)[1]==1)]))
            else:
                first_flag = True
                for roi in rois.split('+'):
                    human_dscalar = f"{datadir}/ROIs/{species}/human.HCP_Parcellation.10k.roi-{roi}.dscalar.nii"
                    values_ = np.concatenate((get_joint_gradient(species,1,'lh')[np.where(get_dscalar(human_dscalar)[0]==1)],
                                             get_joint_gradient(species,1,'rh')[np.where(get_dscalar(human_dscalar)[1]==1)]))
                    if first_flag:
                        values = values_
                        first_flag = False
                    else:
                        values = np.concatenate((values,values_))
        
        elif species == 'marmoset':
            if isinstance(rois, str) and '+' not in rois:
                roi = rois
                marm_dscalar = f"{datadir}/ROIs/{species}/surfFS.MBM_cortex_vPaxinos.10k.roi-{roi}.dscalar.nii"
                values = np.concatenate((get_joint_gradient(species,1,'lh')[np.where(get_dscalar(marm_dscalar)[0]==1)],
                                         get_joint_gradient(species,1,'rh')[np.where(get_dscalar(marm_dscalar)[1]==1)]))
            else:
                first_flag = True
                for roi in rois.split('+'):
                    marm_dscalar = f"{datadir}/ROIs/{species}/surfFS.MBM_cortex_vPaxinos.10k.roi-{roi}.dscalar.nii"
                    values_ = np.concatenate((get_joint_gradient(species,1,'lh')[np.where(get_dscalar(marm_dscalar)[0]==1)],
                                              get_joint_gradient(species,1,'rh')[np.where(get_dscalar(marm_dscalar)[1]==1)]))
                    if first_flag:
                        values = values_
                        first_flag = False
                    else:
                        values = np.concatenate((values,values_))
                        
        else:
            NotImplemented
        
        return values
        
    # human-related functions
    if species == 'human' and option == 'DMN':
        human_dscalar = f"{datadir}/ROIs/human/Ji_2019_human_Default.dscalar.nii"
        values = np.concatenate((get_joint_gradient(species,1,'lh')[np.where(get_dscalar(human_dscalar)[0]==1)],
                                 get_joint_gradient(species,1,'rh')[np.where(get_dscalar(human_dscalar)[1]==1)]))
        return values
    elif species == 'human' and option == 'DMN_no_mPFC':
        roi_list = "PGi+PGs+8Ad+8Av+7m+31pd+31pv+v23ab+POS1+d23ab+31a+23d"
        return get_roi("human",roi_list)
    elif species == 'human' and option != "YEO" and option != "handdrawn":
        return get_roi("human",option)
    # marmoset-related functions
    elif species == 'marmoset' and option == "strict":
        roi_list = "dlPFC+PCC+PPC" # determined via ICA
        return get_roi("marmoset",roi_list)
    elif species == 'marmoset' and option == "lenient":
        roi_list = "A8aD+A6DR+PGM+A23a+A23b+A29d+A30+LIP+MIP+AIP+VIP" # INCLUDE ALL ROIS EXCLUDING mPFC
        return get_roi("marmoset",roi_list)
    elif species == 'marmoset' and option != "lenient" and option != "strict":
        return get_roi("marmoset",option)
    else:
        NotImplemented
```

###### Numerical analysis of node gradient 01 values

```python
from scipy.stats import mannwhitneyu


def cohend(d1, d2):
    # calculate the size of samples
    n1, n2 = len(d1), len(d2)
    # calculate the variance of the samples
    s1, s2 = np.var(d1, ddof=1), np.var(d2, ddof=1)
    # calculate the pooled standard deviation
    s = np.sqrt(((n1 - 1) * s1 + (n2 - 1) * s2) / (n1 + n2 - 2))
    # calculate the means of the samples
    u1, u2 = np.mean(d1), np.mean(d2)
    # calculate the effect size
    return (u1 - u2) / s

print(f"""Human DMN core:
Median: {np.median(human_DMN_2)}
Mean: {np.mean(human_DMN_2)}
St. Dev: {np.std(human_DMN_2)}\n""")

print(f"""Marmoset DMN core:
Median: {np.median(marm_DMN_2)}
Mean: {np.mean(marm_DMN_2)}
St. Dev: {np.std(marm_DMN_2)}""")

print(f"""
Mean difference: {np.abs(np.mean(human_DMN_2) - np.mean(marm_DMN_2))}
Statistics (mannwhitneyu): {mannwhitneyu(human_DMN_2,marm_DMN_2)}""")

roi = 'dlPFC'; marm_dlPFC = get_gradient_distribution('marmoset',roi)
print(f"""
Marmoset dlPFC, mean: {np.mean(marm_dlPFC)}
Marmoset dlPFC, stdev: {np.std(marm_dlPFC)}\n""")

marm_roi_list = "dlPFC+PCC+PPC"
human_roi_list = "dlPFC+PCC+PPC"

manual_order = [(0,3),
                (0,4),
                (0,5),
                (1,3),
                (1,4),
                (1,5),
                (2,3),
                (2,4),
                (2,5),
                (0,1),
                (0,2),
                (1,0),
                (1,2),
                (2,0),
                (2,1),
                (3,4),
                (3,5),
                (4,3),
                (4,5),
                (5,3),
                (5,4)]

effect_size = np.zeros((6,6))
pvals = np.zeros((6,6))

counter = 0
for marm_roi in marm_roi_list.split('+'):
    for human_roi in human_roi_list.split('+'):
        grad1 = get_gradient_distribution('marmoset',marm_roi)
        grad2 = get_gradient_distribution('human',human_roi)
        pvals[manual_order[counter][0],manual_order[counter][1]] = mannwhitneyu(grad1,grad2)[1]
        effect_size[manual_order[counter][0],manual_order[counter][1]] = cohend(grad1,grad2)
        counter += 1
for marm_roi in marm_roi_list.split('+'):
    for human_roi in marm_roi_list.split('+'):
        if marm_roi == human_roi:
            continue
        grad1 = get_gradient_distribution('marmoset',marm_roi)
        grad2 = get_gradient_distribution('marmoset',human_roi)
        pvals[manual_order[counter][0],manual_order[counter][1]] = mannwhitneyu(grad1,grad2)[1]
        effect_size[manual_order[counter][0],manual_order[counter][1]] = cohend(grad1,grad2)
        counter += 1
for marm_roi in human_roi_list.split('+'):
    for human_roi in human_roi_list.split('+'):
        if marm_roi == human_roi:
            continue
        grad1 = get_gradient_distribution('human',marm_roi)
        grad2 = get_gradient_distribution('human',human_roi)
        pvals[manual_order[counter][0],manual_order[counter][1]] = mannwhitneyu(grad1,grad2)[1]
        effect_size[manual_order[counter][0],manual_order[counter][1]] = cohend(grad1,grad2)
        counter += 1
        
effect_size*=np.triu(np.ones((6,6)),)
pvals*=np.triu(np.ones((6,6)),)
```

# Heatmap showing effect-size

```python
import matplotlib
def heatmap(data, row_labels, col_labels, ax=None,
            cbar_kw={}, cbarlabel="", **kwargs):
    """
    Create a heatmap from a numpy array and two lists of labels.

    Parameters
    ----------
    data
        A 2D numpy array of shape (N, M).
    row_labels
        A list or array of length N with the labels for the rows.
    col_labels
        A list or array of length M with the labels for the columns.
    ax
        A `matplotlib.axes.Axes` instance to which the heatmap is plotted.  If
        not provided, use current axes or create a new one.  Optional.
    cbar_kw
        A dictionary with arguments to `matplotlib.Figure.colorbar`.  Optional.
    cbarlabel
        The label for the colorbar.  Optional.
    **kwargs
        All other arguments are forwarded to `imshow`.
    """

    if not ax:
        ax = plt.gca()

    # Plot the heatmap
    im = ax.imshow(data, **kwargs)

    # Create colorbar
    cbar = ax.figure.colorbar(im, ax=ax,pad=0.2, **cbar_kw)
    cbar.ax.set_ylabel(cbarlabel, rotation=-90, va="bottom",fontsize=20)
    cbar.ax.tick_params(labelsize=20,rotation=-90)

    # We want to show all ticks...
    ax.set_xticks(np.arange(data.shape[1]))
    ax.set_yticks(np.arange(data.shape[0]))
    ax.yaxis.tick_right()
    # ... and label them with the respective list entries.
    ax.set_xticklabels(col_labels,fontsize=15)
    ax.set_yticklabels(row_labels,fontsize=15)

    # Let the horizontal axes labeling appear on top.
    ax.tick_params(top=True, bottom=False,
                   labeltop=True, labelbottom=False)


    # Rotate the tick labels and set their alignment.
    plt.setp(ax.get_xticklabels(), rotation=90)

    # Turn spines off and create white grid.
    ax.spines[:].set_visible(False)

    ax.set_xticks(np.arange(data.shape[1]+1)-.5, minor=True)
    ax.set_yticks(np.arange(data.shape[0]+1)-.5, minor=True)
    ax.grid(which="minor", color="w", linestyle='-', linewidth=3)
    ax.tick_params(which="minor", bottom=False, left=False,right=False)

    return im, cbar


def annotate_heatmap(im, data=None, valfmt="{x:.2f}",
                     textcolors=("black", "white"),
                     threshold=None, pval_bin=None,**textkw):
    """
    A function to annotate a heatmap.

    Parameters
    ----------
    im
        The AxesImage to be labeled.
    data
        Data used to annotate.  If None, the image's data is used.  Optional.
    valfmt
        The format of the annotations inside the heatmap.  This should either
        use the string format method, e.g. "$ {x:.2f}", or be a
        `matplotlib.ticker.Formatter`.  Optional.
    textcolors
        A pair of colors.  The first is used for values below a threshold,
        the second for those above.  Optional.
    threshold
        Value in data units according to which the colors from textcolors are
        applied.  If None (the default) uses the middle of the colormap as
        separation.  Optional.
    **kwargs
        All other arguments are forwarded to each call to `text` used to create
        the text labels.
    """

    if not isinstance(data, (list, np.ndarray)):
        data = im.get_array()

    # Normalize the threshold to the images color range.
    if threshold is not None:
        threshold = im.norm(threshold)
    else:
        threshold = im.norm(data.max())/2.

    # Set default alignment to center, but allow it to be
    # overwritten by textkw.
    kw = dict(horizontalalignment="center",
              verticalalignment="center")
    kw.update(textkw)

    # Get the formatter in case a string is supplied
    if isinstance(valfmt, str):
        valfmt = matplotlib.ticker.StrMethodFormatter(valfmt)

    # Loop over the data and create a `Text` for each "pixel".
    # Change the text's color depending on the data.
    texts = []
    for i in range(data.shape[0]):
        for j in range(data.shape[1]):
            if i<j:
                if pval_bin[i,j] == 1:
                    kw.update(fontsize=12,weight='bold',color=textcolors[int(im.norm(data[i, j]) > threshold)])
                    text = im.axes.text(j, i, "*", **kw)
                if pval_bin[i,j] == 2:
                    kw.update(fontsize=12,weight='bold',color=textcolors[int(im.norm(data[i, j]) > threshold)])
                    text = im.axes.text(j, i, "**", **kw)
                if pval_bin[i,j] == 3:
                    kw.update(fontsize=12,weight='bold',color=textcolors[int(im.norm(data[i, j]) > threshold)])
                    text = im.axes.text(j, i, "***", **kw)
                texts.append(text)

    return texts

fig,ax = plt.subplots(figsize=(5,5),dpi=200)

row_labels = ['dlPFC','PCC','PPC']*2
column_labels = ['dlPFC','PCC','PPC']*2
masked_array = np.ma.array(effect_size, mask=effect_size==0)
cmap = matplotlib.cm.Reds
cmap.set_bad('white',1.)

im,cbar = heatmap(np.abs(masked_array),row_labels,row_labels,ax=ax,
                 cmap=cmap, cbarlabel = 'Absolute Cohen\'s d',
                 vmin=0,vmax=5.)

pval_bin = 1*(pvals<(.01/15)) + 1*(pvals<(.001/15)) + 1*(pvals<(.0001/15))
texts = annotate_heatmap(im,pval_bin=pval_bin)

fig.tight_layout()
plt.show()

plt.savefig(os.path.join('figures',f"DMN_effect_size_heatmap.jpg"),dpi=1080,bbox_inches='tight')
```

```python
import pickle 
L_dconn = "/mnt/WD10TB/datasets/data_view_crossspecies/marmoset/dconn/surfFS.lh.graymid.geodesic.10k.dconn.nii"
R_dconn = "/mnt/WD10TB/datasets/data_view_crossspecies/marmoset/dconn/surfFS.rh.graymid.geodesic.10k.dconn.nii"
marmoset_roi_list = "/home/geoff/Desktop/Projects/InProgress/ConnectomeEmbeddings/notebooks/notebook_data/2_DMN_nodes/ROIs/marmoset/Paxinos_roi_list.txt"
human_roi_list = "/home/geoff/Desktop/Projects/InProgress/ConnectomeEmbeddings/notebooks/notebook_data/2_DMN_nodes/ROIs/human/HCP_roi_list.txt"
def get_marmoset_geodesic_distance(dscalar,datadir="notebook_data/2_DMN_nodes",marmoset_roi_list=marmoset_roi_list,human_roi_list=human_roi_list,L_dconn=L_dconn,R_dconn=R_dconn):
    from scipy.stats import mannwhitneyu
    
    print(f"ANCHOR ROI:{dscalar}")
    # Get geodesic file (20484,)
    L_dconn = nib.load(L_dconn).get_fdata()
    R_dconn = nib.load(R_dconn).get_fdata()
    L_dscalar,R_dscalar = get_dscalar(dscalar)
    L_mask_coords = np.where(L_dscalar==1)[0]
    R_mask_coords = np.where(R_dscalar==1)[0]
    geodesic_dist = np.concatenate((L_dconn[L_mask_coords,:].mean(0),
                                    R_dconn[R_mask_coords,:].mean(0)))
    # Get gradient file (20484,)
    gradient1 = {}
    for species in ['marmoset','human']:
        gradient1[species] = np.concatenate((get_joint_gradient(species,1,'lh'),
                                             get_joint_gradient(species,1,'rh')))
    
    # Loop through marmoset ROIs
    species = 'marmoset'
    f = open(marmoset_roi_list)
    avg_geodesic_dist_list = []
    gradient_values_list = []
    roi_list = []
    for roi in f.readlines():
        roi = roi.strip()
        marm_dscalar = f"{datadir}/ROIs/{species}/surfFS.MBM_cortex_vPaxinos.10k.roi-{roi}.dscalar.nii"
        L_coords = np.where(get_dscalar(marm_dscalar)[0]==1)
        R_coords = np.where(get_dscalar(marm_dscalar)[1]==1)
        # Save average geodesic distance from `dscalar`
        geodesic_mean = np.concatenate((geodesic_dist[:10242][L_coords],
                                        geodesic_dist[10242:][R_coords])).mean()
        avg_geodesic_dist_list.append(geodesic_mean)
        # Save list of gradient values from atlas ROIs
        gradient_values = np.concatenate((gradient1[species][:10242][L_coords],
                                          gradient1[species][10242:][R_coords]))
        gradient_values = gradient_values[np.where(gradient_values>0)] # Remove zeroes, this should only affect one vertex
        gradient_values_list.append(gradient_values)
        # Save roi name
        roi_list.append(roi)
    f.close()
    # Reorder marmoset rois
    rois_reordered = []
    gradient_values_reordered = []
    pvalues_reordered = []
    for idx in np.argsort(avg_geodesic_dist_list):
        rois_reordered.append(roi_list[idx])
        gradient_values_reordered.append(gradient_values_list[idx])
        pvalues_reordered.append(mannwhitneyu(gradient_values_reordered[0],gradient_values_list[idx])[1]*len(roi_list))
    
    def link_roi_to_network(L_coords,R_coords):
        rsns = ['Auditory','Cingulo-Opercular',
                'Default','Dorsal-Attention',
                'Frontoparietal','Language',
                'Orbito-Affective','Posterior-Multimodal',
                'Somatomotor','Ventral-Multimodal',
                'Visual2','Visual']
        baseline = 0
        for rsn in rsns:
            rsn_dscalar = f"/home/geoff/Desktop/Projects/InProgress/ConnectomeEmbeddings/notebooks/notebook_data/2_DMN_nodes/ROIs/human/Ji_2019_human_{rsn}.dscalar.nii"
            L_dscalar,R_dscalar = get_dscalar(rsn_dscalar)
            mask_sum = np.concatenate((L_dscalar[L_coords],
                                         R_dscalar[R_coords])).sum()
            if mask_sum > baseline:
                baseline = mask_sum
                network = rsn
        return network
    
    # Loop through human ROIs
    species = 'human'
    f = open(human_roi_list)
    human_roi_list = []
    human_gradient_values_list = []
    human_network_list = []
    human_pvalues = []
    for roi in tqdm(f.readlines()):
        roi = roi.strip()
        human_dscalar = f"{datadir}/ROIs/{species}/human.HCP_Parcellation.10k.roi-{roi}.dscalar.nii"
        L_coords = np.where(get_dscalar(human_dscalar)[0]==1)
        R_coords = np.where(get_dscalar(human_dscalar)[1]==1)
        # Save list of gradient values from atlas ROIs
        gradient_values = np.concatenate((gradient1[species][:10242][L_coords],
                                          gradient1[species][10242:][R_coords]))
        gradient_values = gradient_values[np.where(gradient_values>0)] # Remove zeroes, this should only affect one vertex
        human_gradient_values_list.append(gradient_values)
        # Get network
        human_network_list.append(link_roi_to_network(L_coords,R_coords))
        # Get pvalues
        human_pvalues.append(mannwhitneyu(gradient_values_reordered[0],gradient_values)[1]*len(human_roi_list))
        # Save roi name
        human_roi_list.append(roi)
    f.close()
    
    return rois_reordered,gradient_values_reordered,pvalues_reordered,human_roi_list,human_gradient_values_list,human_network_list,human_pvalues

pkldir=f"notebook_data/2_DMN_nodes/pkl/main"
fn = os.path.join(pkldir,f"marmoset_dmn_node_analysis.pkl")
if os.path.exists(fn):
    print(f"{fn} exists.")
    f = open(fn, 'rb')
    data = pickle.load(f)
    labels = data[0];values = data[1];pvalues = data[2];human_labels = data[3]
    human_values = data[4];human_networks = data[5];human_pvalues = data[6]
else:
    print(f"{fn} does not exist.")
    # Marmoset
    labels = {}
    values = {}
    pvalues = {}
    # Human
    human_labels = {}
    human_values = {}
    human_networks = {}
    human_pvalues = {}
    # Loop through Liu 2020 DMN nodes (defined by peaks)
    for roi in ['dlPFC','PCC','PPC']:
        dscalar = f"/home/geoff/Desktop/Projects/InProgress/ConnectomeEmbeddings/notebooks/notebook_data/2_DMN_nodes/ROIs/marmoset/surfFS.MBM_cortex_vPaxinos.10k.roi-{roi}.dscalar.nii"
        labels[roi], values[roi], pvalues[roi],human_labels[roi],human_values[roi],human_networks[roi],human_pvalues[roi] = get_marmoset_geodesic_distance(dscalar)
    marmoset_dmn_node_data = [labels,values,pvalues,human_labels,human_values,human_networks,human_pvalues]
    print(f"SAVING: {fn}")
    f=open(fn,"wb")
    pickle.dump(marmoset_dmn_node_data,f,2)
    del marmoset_dmn_node_data
```

```python
def get_roi_name(roi,datadir="notebook_data/2_DMN_nodes/ROIs/human/20484"):
    basename = f"human.HCP_Parcellation.10k.roi-{roi}.dscalar.nii"
    return os.path.join(datadir,basename)

def get_cifti_add_cmd(dscalar_out,roi,save_roi):
    species = 'human'
    dscalar_gradient = os.path.join(jointgrad_dir,f"joint_eigenmap_{species}.dscalar.nii")
    roi_list = []
    roi_list.append(get_roi_name(roi))
    for roi in save_roi[roi]:
        roi_list.append(get_roi_name(roi))
    if len(roi_list) == 1:
        return 1
    cmds = []
    for ix,roi in enumerate(roi_list):
        if ix == 1:
            cmd = f"wb_command -cifti-math 'X+Y' {dscalar_out} -var X {roi} -var Y {roi}"
        elif ix > 1:
            cmd = f"wb_command -cifti-math 'X+Y' {dscalar_out} -var X {dscalar_out} -var Y {roi}"
        else:
            continue
        cmds.append(cmd)

    cmd = f"wb_command -cifti-math 'X>0' {dscalar_out} -var X {dscalar_out}"
    cmds.append(cmd)
    cmd = f"wb_command -cifti-math 'X*Y' {dscalar_out} -var X {dscalar_out} -var Y {dscalar_gradient} -select 1 2"
    cmds.append(cmd)
    
    for cmd in cmds:
        os.system(cmd)
        
def convert_10k_to_32k(species,dscalar_10k):
    
    """
    INPUTS into wb_command -cifti-resample
    """
    dscalar_32k = dscalar_10k.replace('dscalar.nii','32k.dscalar.nii')
    if species == "marmoset":
        template = "/mnt/WD10TB/templates/Marmoset_Brain_Mappping_v3.0.1/Marmoset_Brain_Mappping_v3.0.1/MBM_v3.0.1/derivatives/surfFS.MBM_cortex_vPaxinos.dscalar.nii"
        lh_sphere_10k = "/mnt/WD10TB/templates/Marmoset_Brain_Mappping_v3.0.1/Marmoset_Brain_Mappping_v3.0.1/MBM_v3.0.1/derivatives/10k/lh.sphere.10k.surf.gii"
        lh_sphere_32k = "/mnt/WD10TB/templates/Marmoset_Brain_Mappping_v3.0.1/Marmoset_Brain_Mappping_v3.0.1/MBM_v3.0.1/derivatives/10k/intermediate/surfFS.lh.sphere.match.surf.gii"
        lh_mid_10k = "/mnt/WD10TB/templates/Marmoset_Brain_Mappping_v3.0.1/Marmoset_Brain_Mappping_v3.0.1/MBM_v3.0.1/derivatives/10k/surfFS.lh.graymid.10k.surf.gii"
        lh_mid_32k = "/mnt/WD10TB/templates/Marmoset_Brain_Mappping_v3.0.1/Marmoset_Brain_Mappping_v3.0.1/MBM_v3.0.1/surfFS.lh.graymid.surf.gii"
        rh_sphere_10k = "/mnt/WD10TB/templates/Marmoset_Brain_Mappping_v3.0.1/Marmoset_Brain_Mappping_v3.0.1/MBM_v3.0.1/derivatives/10k/rh.sphere.10k.surf.gii"
        rh_sphere_32k = "/mnt/WD10TB/templates/Marmoset_Brain_Mappping_v3.0.1/Marmoset_Brain_Mappping_v3.0.1/MBM_v3.0.1/derivatives/10k/intermediate/surfFS.rh.sphere.match.surf.gii"
        rh_mid_10k = "/mnt/WD10TB/templates/Marmoset_Brain_Mappping_v3.0.1/Marmoset_Brain_Mappping_v3.0.1/MBM_v3.0.1/derivatives/10k/surfFS.rh.graymid.10k.surf.gii"
        rh_mid_32k = "/mnt/WD10TB/templates/Marmoset_Brain_Mappping_v3.0.1/Marmoset_Brain_Mappping_v3.0.1/MBM_v3.0.1/surfFS.rh.graymid.surf.gii"
    if species == "human":
        template = "/mnt/WD10TB/templates/surfaces/S1200.MyelinMap_BC_MSMAll.32k_fs_LR.dscalar.nii"
        lh_sphere_10k = "/mnt/WD10TB/templates/standard_mesh_atlases/L.sphere.10k_fs_LR.surf.gii"
        lh_sphere_32k = "/mnt/WD10TB/templates/standard_mesh_atlases/L.sphere.32k_fs_LR.surf.gii"
        lh_mid_10k = "/mnt/WD10TB/templates/surfaces/S1200.L.midthickness_MSMAll.10k_fs_LR.surf.gii"
        lh_mid_32k = "/mnt/WD10TB/templates/surfaces/S1200.L.midthickness_MSMAll.32k_fs_LR.surf.gii"
        rh_sphere_10k = "/mnt/WD10TB/templates/standard_mesh_atlases/R.sphere.10k_fs_LR.surf.gii"
        rh_sphere_32k = "/mnt/WD10TB/templates/standard_mesh_atlases/R.sphere.32k_fs_LR.surf.gii"
        rh_mid_10k = "/mnt/WD10TB/templates/surfaces/S1200.R.midthickness_MSMAll.10k_fs_LR.surf.gii"
        rh_mid_32k = "/mnt/WD10TB/templates/surfaces/S1200.R.midthickness_MSMAll.32k_fs_LR.surf.gii"

    cmd = f"wb_command -cifti-resample {dscalar_10k} COLUMN {template} COLUMN ADAP_BARY_AREA CUBIC {dscalar_32k} -left-spheres {lh_sphere_10k} {lh_sphere_32k} -left-area-surfs {lh_mid_10k} {lh_mid_32k} -right-spheres {rh_sphere_10k} {rh_sphere_32k} -right-area-surfs {rh_mid_10k} {rh_mid_32k} -surface-largest"
    os.system(cmd)
    
    return dscalar_32k

for roi in save_roi.keys():
    get_cifti_add_cmd(f"notebook_data/2_DMN_nodes/visualization/human/similar_gradient_values_to_roi-{roi}.dscalar.nii",roi,human_save_roi)
    convert_10k_to_32k('human',f"notebook_data/2_DMN_nodes/visualization/human/similar_gradient_values_to_roi-{roi}.dscalar.nii")
```
