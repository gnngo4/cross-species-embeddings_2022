### Node embedding with Laplacian Eigenmaps with scikit-learn [primary eigenvector]

```python
def adjacency_to_graph_laplacian(adj):
    n_nodes = adj.shape[0]
    rowsum = np.array(adj.sum(1))
    D_inv_sqrt = np.power(rowsum,-.5).flatten()
    D_inv_sqrt[np.isinf(D_inv_sqrt)] = 0.
    D_inv_sqrt = np.diag(D_inv_sqrt)
    L_sym = np.eye(n_nodes) - np.dot(np.dot(D_inv_sqrt,adj),D_inv_sqrt)
    return L_sym

# PCA Visualization
def pca_le(X):
    from sklearn.decomposition import PCA

    pca = PCA()
    pca.fit(X)
    transformed = pca.transform(X)
    plt.scatter(transformed[:,0],transformed[:,1],color=colors,lw=1,edgecolor='k')
    plt.xlabel('PCA1'); plt.ylabel('PCA2')

# Convert barbell graph to graph Laplacian
adj = np.array(nx.adjacency_matrix(G).todense())
L_sym = adjacency_to_graph_laplacian(adj)
    
embedding_machine = LaplacianEigenmaps(L_sym,_)
embedding_machine.create_embedding()
```

```python
fig = plt.figure(figsize=(12,8),dpi=100)

ax = fig.add_subplot(2,3,1)
plt.title("Graph")
nodes = nx.draw_networkx(G,pos=pos,
                         node_size=150,node_color=colors,with_labels=False,
                         width=1.,edgecolors='k')

ax = fig.add_subplot(2,3,4)
eig_id=1
plt.title(f"Eigenmap: {eig_id}")
nodes = nx.draw_networkx(G,pos=pos,
                         node_size=150,node_color=embedding_machine.eigen_vectors[:,eig_id],
                         cmap='coolwarm',with_labels=False,
                         width=1.,edgecolors='k')

ax = fig.add_subplot(2,3,5)
eig_id=2
plt.title(f"Eigenmap: {eig_id}")
nodes = nx.draw_networkx(G,pos=pos,
                         node_size=150,node_color=embedding_machine.eigen_vectors[:,eig_id],
                         cmap='coolwarm',with_labels=False,
                         width=1.,edgecolors='k')

ax = fig.add_subplot(2,3,6)
eig_id=3
plt.title(f"Eigenmap: {eig_id}")
nodes = nx.draw_networkx(G,pos=pos,
                         node_size=150,node_color=embedding_machine.eigen_vectors[:,eig_id],
                         cmap='coolwarm',with_labels=False,
                         width=1.,edgecolors='k')

ax = fig.add_subplot(2,3,2)
eig_id=2
plt.title("Laplacian Eigenmap Embedding")
plt.scatter(embedding_machine.eigen_vectors[:,1],embedding_machine.eigen_vectors[:,2],color=colors,edgecolor='k')
plt.xlabel('LE1'); plt.ylabel('LE2')

ax = fig.add_subplot(2,3,3)
plt.title('PCA of Laplacian Eigenmap Embedding')
pca_le(embedding_machine.eigen_vectors)

plt.tight_layout()
```

### Node embedding with GraphWave

```python
# PCA Visualization
def pca_struc_embedding(real,imag,add_noise=True):
    from sklearn.decomposition import PCA

    structural_embedding = np.concatenate((real,imag),axis=1)
    pca = PCA()
    pca.fit(structural_embedding)
    transformed = pca.transform(structural_embedding)
    if add_noise:
        transformed += np.random.normal(0,.1,size=(transformed.shape)) # Add noise to visualize overlap
    plt.scatter(transformed[:,0],transformed[:,1],color=colors,lw=1,edgecolor='k')
    plt.xlabel('PCA1'); plt.ylabel('PCA2')

class Namespace:
    def __init__(self,**kwargs):
        self.__dict__.update(kwargs)

settings = Namespace(gw_range_nodes = '',
                     step_size = 1,
                     sample_number = 128,
                     heat_coefficient = 10)
        
embedding_machine = GraphWave(L_sym,settings)
embedding_machine.create_embedding()

print(embedding_machine.real_and_imaginary.shape)
```

```python
X = embedding_machine.real_and_imaginary
n_nodes = X.shape[0]
real = X.real
imag = X.imag

fig = plt.figure(figsize=(12,8),dpi=100)

ax = fig.add_subplot(2,3,1)
plt.title('Graph')
nodes = nx.draw_networkx(G,pos=pos,
                         node_size=150,node_color=colors,with_labels=True,
                         width=1.,edgecolors='k')

ax = fig.add_subplot(2,3,4)
plt.title('GraphWave Embedding')
for i in range(n_nodes):
    plt.plot(real[i,:],imag[i,:],c=colors[i],
             alpha=.9)
plt.xlabel('Re.'); plt.ylabel('Im.')

ax = fig.add_subplot(2,3,2)
plt.title('PCA of GraphWave Embedding [with noise]')
pca_struc_embedding(real,imag)

ax = fig.add_subplot(2,3,3)
plt.title('PCA of GraphWave Embedding [no noise]')
pca_struc_embedding(real,imag,add_noise=False)
    
ax = fig.add_subplot(2,3,5)
nodeX=5;plt.plot(real[nodeX,:],imag[nodeX,:],c='red',lw=4.)
nodeY=1;plt.plot(real[nodeY,:],imag[nodeY,:],c='blue')
plt.title(f"node[{nodeX}]:red\nnode[{nodeY}]:blue")
plt.xlabel('Re.'); plt.ylabel('Im.')

ax = fig.add_subplot(2,3,6)
nodeX=10;plt.plot(real[nodeX,:],imag[nodeX,:],c='red',lw=4.)
nodeY=1;plt.plot(real[nodeY,:],imag[nodeY,:],c='blue')
plt.title(f"node[{nodeX}]:red\nnode[{nodeY}]:blue")
plt.xlabel('Re.'); plt.ylabel('Im.')

plt.tight_layout()
```
