```python
def get_joint_gradient(species,gradient_id,hemi,jointgrad_dir=jointgrad_dir,sing_shell=sing_shell):
    
    def read_dscalar(dscalar,hemi):
        # Set-up tmpdir
        tmpdir = tempfile.TemporaryDirectory()
        # Read data
        left_func = f"{os.path.join(tmpdir.name,'left_func.func.gii')}"
        right_func = f"{os.path.join(tmpdir.name,'right_func.func.gii')}"
        cmd = f"{sing_shell} wb_command -cifti-separate {dscalar} COLUMN -metric CORTEX_LEFT {left_func} -metric CORTEX_RIGHT {right_func}"
        os.system(cmd)
        if hemi == 'lh':
            read_func = left_func
        elif hemi == 'rh':
            read_func = right_func
        else:
            NotImplemented
        
        return nib.load(read_func).agg_data()
        
    gradient_path = os.path.join(jointgrad_dir,f"joint_eigenmap_{species}.dscalar.nii")
    data = read_dscalar(gradient_path,hemi)
    data = data[gradient_id]
    
    return data

def get_dscalar(dscalar,sing_shell=sing_shell):
    
    def read_dscalar(dscalar,hemi):
        # Set-up tmpdir
        tmpdir = tempfile.TemporaryDirectory()
        # Read data
        left_func = f"{os.path.join(tmpdir.name,'left_func.func.gii')}"
        right_func = f"{os.path.join(tmpdir.name,'right_func.func.gii')}"
        cmd = f"{sing_shell} wb_command -cifti-separate {dscalar} COLUMN -metric CORTEX_LEFT {left_func} -metric CORTEX_RIGHT {right_func}"
        os.system(cmd)
        if hemi == 'lh':
            read_func = left_func
        elif hemi == 'rh':
            read_func = right_func
        else:
            NotImplemented
        
        return nib.load(read_func).agg_data()
        
    data_lh = read_dscalar(dscalar,'lh')
    data_rh = read_dscalar(dscalar,'rh')
    
    return data_lh, data_rh


def get_gradient_distribution(species,option,datadir="notebook_data/2_DMN_nodes"):
    """
    Function to grab gradient values from the somatomotor-DMN gradient
    """
    def get_roi(species,rois):
        """
        * rois is '+'-delimited
        Human ROIs: dlPFC, PCC, PPC
        Marmoset ROIs: dlPFC, PCC, PPC, mPFC [A32,A10,A32V]
        """
        if species == 'human':
            if isinstance(rois, str) and '+' not in rois:
                roi = rois
                human_dscalar = f"{datadir}/ROIs/{species}/human.HCP_Parcellation.10k.roi-{roi}.dscalar.nii"
                values = np.concatenate((get_joint_gradient(species,1,'lh')[np.where(get_dscalar(human_dscalar)[0]==1)],
                                         get_joint_gradient(species,1,'rh')[np.where(get_dscalar(human_dscalar)[1]==1)]))
            else:
                first_flag = True
                for roi in rois.split('+'):
                    human_dscalar = f"{datadir}/ROIs/{species}/human.HCP_Parcellation.10k.roi-{roi}.dscalar.nii"
                    values_ = np.concatenate((get_joint_gradient(species,1,'lh')[np.where(get_dscalar(human_dscalar)[0]==1)],
                                             get_joint_gradient(species,1,'rh')[np.where(get_dscalar(human_dscalar)[1]==1)]))
                    if first_flag:
                        values = values_
                        first_flag = False
                    else:
                        values = np.concatenate((values,values_))
        
        elif species == 'marmoset':
            if isinstance(rois, str) and '+' not in rois:
                roi = rois
                marm_dscalar = f"{datadir}/ROIs/{species}/surfFS.MBM_cortex_vPaxinos.10k.roi-{roi}.dscalar.nii"
                values = np.concatenate((get_joint_gradient(species,1,'lh')[np.where(get_dscalar(marm_dscalar)[0]==1)],
                                         get_joint_gradient(species,1,'rh')[np.where(get_dscalar(marm_dscalar)[1]==1)]))
            else:
                first_flag = True
                for roi in rois.split('+'):
                    marm_dscalar = f"{datadir}/ROIs/{species}/surfFS.MBM_cortex_vPaxinos.10k.roi-{roi}.dscalar.nii"
                    values_ = np.concatenate((get_joint_gradient(species,1,'lh')[np.where(get_dscalar(marm_dscalar)[0]==1)],
                                              get_joint_gradient(species,1,'rh')[np.where(get_dscalar(marm_dscalar)[1]==1)]))
                    if first_flag:
                        values = values_
                        first_flag = False
                    else:
                        values = np.concatenate((values,values_))
                        
        else:Spatial correlation analysesÂ¶
            NotImplemented
        
        return values
        
    # human-related functions
    if species == 'human' and option == 'DMN':
        human_dscalar = f"{datadir}/ROIs/human/Ji_2019_human_Default.dscalar.nii"
        values = np.concatenate((get_joint_gradient(species,1,'lh')[np.where(get_dscalar(human_dscalar)[0]==1)],
                                 get_joint_gradient(species,1,'rh')[np.where(get_dscalar(human_dscalar)[1]==1)]))
        return values
    elif species == 'human' and option == 'DMN_no_mPFC':
        roi_list = "PGi+PGs+8Ad+8Av+7m+31pd+31pv+v23ab+POS1+d23ab+31a+23d"
        return get_roi("human",roi_list)
    elif species == 'human' and option != "YEO" and option != "handdrawn":
        return get_roi("human",option)
    # marmoset-related functions
    elif species == 'marmoset' and option == "strict":
        roi_list = "dlPFC+PCC+PPC" # INCLUDE ONLY PEAK ROIS FOR EACH DMN NODE (dlPFC/PCC/PPC) according to Liu et al 2019
        return get_roi("marmoset",roi_list)
    elif species == 'marmoset' and option == "lenient":
        roi_list = "A8aD+A6DR+PGM+A23a+A23b+A29d+A30+LIP+MIP+AIP+VIP" # INCLUDE ALL ROIS EXCLUDING mPFC
        return get_roi("marmoset",roi_list)
    elif species == 'marmoset' and option != "lenient" and option != "strict":
        return get_roi("marmoset",option)
    else:
        NotImplemented
        
def get_rfmri_data(species,rfmri_dir):
    
    def summarize_fn(dtseries,species):
        if species == 'human':
            dtseries = dtseries.split('/')[-1].replace('.wmcsf_reg.s6.cortex.dtseries.nii','').replace('.rfMRI_','_')
            return dtseries
        elif species == 'marmoset':
            dtseries = dtseries.replace('.rfMRI.wmcsf_reg.s1.5.cortex.dtseries.nii','')
            return dtseries
        else:
            NotImplemented
        return dtseries

    all_data = os.listdir(rfmri_dir[species])
    data = {}
    data['id'] = []
    data['lh'] = []
    data['rh'] = []
    for ix, dtseries in tqdm(enumerate(all_data)):
        fn = summarize_fn(dtseries,species)
        data['id'].append(fn)
        
        # COMMENTOUT
        """dtseries = os.path.join(rfmri_dir[species],dtseries)
        lh_data, rh_data = get_dscalar(dtseries)
        n_tp = len(lh_data)
        n_coords = lh_data[0].shape[0]
        
        data_temp = np.zeros((n_coords,n_tp))
        for ix_,row in enumerate(lh_data):
            data_temp[:,ix_] = row
        data_temp = (data_temp - data_temp.mean(1,keepdims=True))/(data_temp.std(1,keepdims=True))
        data['lh'].append(data_temp)
        data_temp = np.zeros((n_coords,n_tp))
        for ix_,row in enumerate(rh_data):
            data_temp[:,ix_] = row    
        data_temp = (data_temp - data_temp.mean(1,keepdims=True))/(data_temp.std(1,keepdims=True))
        data['rh'].append(data_temp)"""
        # COMMENTOUT
        
    return data
    
def get_timeseries(roi,species,rfmri_data,datadir="notebook_data/2_DMN_nodes"):
    
    print(roi,species)
    
    def get_roi_coords(species,roi):
        
        """
        Human ROIs: dlPFC [8Ad,8Av], PCC [7m,31pd,31pv,v23ab,POS1,d23ab,31a,23d],
                    PPC [PGi,PGs]
        Marmoset ROIs: dlPFC, PCC, PPC, mPFC [A32,A10,A32V]
        """
        if species == 'human':
            human_dscalar = f"{datadir}/ROIs/{species}/human.HCP_Parcellation.10k.roi-{roi}.dscalar.nii"
            lh_coords = np.where(get_dscalar(human_dscalar)[0]==1)
            rh_coords = np.where(get_dscalar(human_dscalar)[1]==1)

        elif species == 'marmoset':
            marm_dscalar = f"{datadir}/ROIs/{species}/surfFS.MBM_cortex_vPaxinos.10k.roi-{roi}.dscalar.nii"
            lh_coords = np.where(get_dscalar(marm_dscalar)[0]==1)
            rh_coords = np.where(get_dscalar(marm_dscalar)[1]==1)
        
        return lh_coords,rh_coords
    
    data = {}
    data['id'] = []
    data['lh'] = []
    data['rh'] = []
    lh_coords, rh_coords = get_roi_coords(species,roi)
    for ix,ts_data in enumerate(rfmri_data['lh']):
        fn = rfmri_data['id'][ix]
        data['id'].append(fn)
        extracted_ts = ts_data[lh_coords]
        data['lh'].append(extracted_ts)
    for ix,ts_data in enumerate(rfmri_data['rh']):
        extracted_ts = ts_data[rh_coords]
        data['rh'].append(extracted_ts)
        
    return data

def get_tsnr_data(roi,species,rfmri_dir,datadir="notebook_data/2_DMN_nodes"):

    def get_roi_coords(species,roi):
        
        """
        Human ROIs: dlPFC [8Ad,8Av], PCC [7m,31pd,31pv,v23ab,POS1,d23ab,31a,23d],
                    PPC [PGi,PGs]
        Marmoset ROIs: dlPFC, PCC, PPC, mPFC [A32,A10,A32V]
        """
        if species == 'human':
            human_dscalar = f"{datadir}/ROIs/{species}/human.HCP_Parcellation.10k.roi-{roi}.dscalar.nii"
            lh_coords = np.where(get_dscalar(human_dscalar)[0]==1)
            rh_coords = np.where(get_dscalar(human_dscalar)[1]==1)

        elif species == 'marmoset':
            marm_dscalar = f"{datadir}/ROIs/{species}/surfFS.MBM_cortex_vPaxinos.10k.roi-{roi}.dscalar.nii"
            lh_coords = np.where(get_dscalar(marm_dscalar)[0]==1)
            rh_coords = np.where(get_dscalar(marm_dscalar)[1]==1)
        
        return lh_coords,rh_coords
    
    def summarize_fn(dtseries,species):
        if species == 'human':
            dtseries = dtseries.split('/')[-1].replace('.wmcsf_reg.s6.cortex.dtseries.nii','').replace('.rfMRI_','_')
            return dtseries
        elif species == 'marmoset':
            dtseries = dtseries.replace('.rfMRI.tsnr.dscalar.nii','')
            return dtseries
        else:
            NotImplemented
        return dtseries

    all_data = os.listdir(rfmri_dir)
    lh_coords, rh_coords = get_roi_coords(species,roi)
    data = {}
    data['id'] = []
    data['lh'] = []
    data['rh'] = []
    for ix, dtseries in tqdm(enumerate(all_data)):
        fn = summarize_fn(dtseries,species)
        data['id'].append(fn)
        dtseries = os.path.join(rfmri_dir,dtseries)
        lh_data, rh_data = get_dscalar(dtseries)
        data['lh'].append(lh_data[lh_coords])
        data['rh'].append(rh_data[rh_coords])
        
    return data

class connectivity_info:
    
    def __init__(self,id_,cortex_ts,dmn_ts,dmn_order):
        self.id = id_
        self.cortex_ts = cortex_ts
        self.dmn_ts = dmn_ts
        self.dmn_order = dmn_order
        self.n_runs = len(self.cortex_ts)
        
    def process(self):
        first_flag = True
        for run in range(self.n_runs):
            if first_flag:
                self.dmn_to_dmn = np.corrcoef(self.dmn_ts[run],self.dmn_ts[run])[:6,:][:,:6]
                self.dmn_to_cortex = np.corrcoef(self.cortex_ts[run],self.dmn_ts[run])[:,-6:][:-6,:]
                first_flag = False
            else:
                self.dmn_to_dmn += np.corrcoef(self.dmn_ts[run],self.dmn_ts[run])[:6,:][:,:6]
                self.dmn_to_cortex += np.corrcoef(self.cortex_ts[run],self.dmn_ts[run])[:,-6:][:-6,:]
        self.dmn_to_dmn /= self.n_runs
        self.dmn_to_cortex /= self.n_runs
        del self.dmn_ts, self.cortex_ts

def condense_data_by_participant(species,species_id,rfmri_data,ts_data):
    
    import pickle
            
    def get_roi_from_ts():
        first_roi_flag = True
        roi_list = []
        for hemi in ['lh','rh']:
            for roi in ts_data.keys():
                roi_list.append(f"{roi} {hemi}")
                if species in roi:
                    if first_roi_flag:
                        roi_ts = ts_data[roi][hemi][ix].mean(0,keepdims=True)
                        first_roi_flag = False
                    else:
                        roi_ts = np.vstack((roi_ts,ts_data[roi][hemi][ix].mean(0,keepdims=True)))
                        
        return roi_ts,roi_list
    
    pkldir=f"notebook_data/2_DMN_nodes/pkl/main"
    id_list = []
    for participant_id in tqdm(species_id):
        id_list.append(participant_id)
        fn = os.path.join(pkldir,f"{participant_id}.pkl")
        if os.path.exists(fn):
            continue
        wholebrain_ts = []
        roi_ts = []
        for ix, search_id in enumerate(rfmri_data['id']):
            if participant_id in search_id:
                wholebrain_ts.append(np.concatenate((rfmri_data['lh'][ix],rfmri_data['rh'][ix]),axis=0))
                roi_ts.append(get_roi_from_ts()[0])
        data_ = connectivity_info(participant_id,wholebrain_ts,roi_ts,get_roi_from_ts()[1])
        data_.process()
        f=open(fn,"wb")
        pickle.dump(data_,f,2)
        
    return id_list

def get_unique_ids(rfmri_human,rfmri_marmoset,species):
    
    if species == 'human':
        ids = list(set([scanid.split('_')[0] for scanid in rfmri_human['id']]))
    elif species == 'marmoset':
        ids = list(set([scanid.split('_')[0] for scanid in rfmri_marmoset['id']]))
        ids = [i for i in ids if 'sub' in i]
        
    return ids
```

```python
import matplotlib
def heatmap(data, row_labels, col_labels, ax=None,
            cbar_kw={}, cbarlabel="", **kwargs):
    """
    Create a heatmap from a numpy array and two lists of labels.

    Parameters
    ----------
    data
        A 2D numpy array of shape (N, M).
    row_labels
        A list or array of length N with the labels for the rows.
    col_labels
        A list or array of length M with the labels for the columns.
    ax
        A `matplotlib.axes.Axes` instance to which the heatmap is plotted.  If
        not provided, use current axes or create a new one.  Optional.
    cbar_kw
        A dictionary with arguments to `matplotlib.Figure.colorbar`.  Optional.
    cbarlabel
        The label for the colorbar.  Optional.
    **kwargs
        All other arguments are forwarded to `imshow`.
    """

    if not ax:
        ax = plt.gca()

    # Plot the heatmap
    im = ax.imshow(data, **kwargs)

    # Create colorbar
    cbar = ax.figure.colorbar(im, ax=ax,pad=0.2, **cbar_kw)
    cbar.ax.set_ylabel(cbarlabel, rotation=-90, va="bottom",fontsize=20)
    cbar.ax.tick_params(labelsize=20,rotation=0)

    # We want to show all ticks...
    ax.set_xticks(np.arange(data.shape[1]))
    ax.set_yticks(np.arange(data.shape[0]))
    ax.yaxis.tick_right()
    # ... and label them with the respective list entries.
    ax.set_xticklabels(col_labels,fontsize=15)
    ax.set_yticklabels(row_labels,fontsize=15)

    # Let the horizontal axes labeling appear on top.
    ax.tick_params(top=True, bottom=False,
                   labeltop=True, labelbottom=False)


    # Rotate the tick labels and set their alignment.
    plt.setp(ax.get_xticklabels(), rotation=90)

    # Turn spines off and create white grid.
    ax.spines[:].set_visible(False)

    ax.set_xticks(np.arange(data.shape[1]+1)-.5, minor=True)
    ax.set_yticks(np.arange(data.shape[0]+1)-.5, minor=True)
    ax.grid(which="minor", color="w", linestyle='-', linewidth=3)
    ax.tick_params(which="minor", bottom=False, left=False,right=False)

    return im, cbar


def annotate_heatmap(im, data=None, valfmt="{x:.2f}",
                     textcolors=("black", "white"),
                     threshold=None, pval_bin=None,**textkw):
    """
    A function to annotate a heatmap.

    Parameters
    ----------
    im
        The AxesImage to be labeled.
    data
        Data used to annotate.  If None, the image's data is used.  Optional.
    valfmt
        The format of the annotations inside the heatmap.  This should either
        use the string format method, e.g. "$ {x:.2f}", or be a
        `matplotlib.ticker.Formatter`.  Optional.
    textcolors
        A pair of colors.  The first is used for values below a threshold,
        the second for those above.  Optional.
    threshold
        Value in data units according to which the colors from textcolors are
        applied.  If None (the default) uses the middle of the colormap as
        separation.  Optional.
    **kwargs
        All other arguments are forwarded to each call to `text` used to create
        the text labels.
    """

    if not isinstance(data, (list, np.ndarray)):
        data = im.get_array()

    # Normalize the threshold to the images color range.
    if threshold is not None:
        threshold = im.norm(threshold)
    else:
        threshold = im.norm(data.max())/2.

    # Set default alignment to center, but allow it to be
    # overwritten by textkw.
    kw = dict(horizontalalignment="center",
              verticalalignment="center")
    kw.update(textkw)

    # Get the formatter in case a string is supplied
    if isinstance(valfmt, str):
        valfmt = matplotlib.ticker.StrMethodFormatter(valfmt)

    # Loop over the data and create a `Text` for each "pixel".
    # Change the text's color depending on the data.
    texts = []
    for i in range(data.shape[0]):
        for j in range(data.shape[1]):
            if i<j:
                kw.update(fontsize=12,weight='bold',color=textcolors[int(im.norm(data[i, j]) > threshold)])
                text = im.axes.text(j, i, f"{data[i,j]:.2f}", **kw)
                texts.append(text)

    return texts

fig,ax = plt.subplots(figsize=(5,5),dpi=200)

conn_diff = get_dmn_to_dmn_mean(human_data) - get_dmn_to_dmn_mean(marmoset_data)
row_labels = ['dlPFC','PCC','PPC']*2
column_labels = ['dlPFC','PCC','PPC']*2
masked_array = np.ma.array(conn_diff, mask=np.tri(conn_diff.shape[0], k=-1)==1)
cmap = matplotlib.cm.Reds
cmap.set_bad('white',1.)

im,cbar = heatmap(masked_array,
                  row_labels,row_labels,ax=ax,
                  cmap=cmap, cbarlabel = "RSFC difference\n(Human-Marmoset)",
                  vmin=0,
                  vmax=.5)

texts = annotate_heatmap(im,fontsize=10)

fig.tight_layout()
plt.show()

plt.savefig(os.path.join('figures',f"Human_subtract_marmoset_conn.jpg"),dpi=1080,bbox_inches='tight')
```

```python
def get_rsfc(marmoset_ids,method="mean"):
    dmn_order = read_pkl(marmoset_ids[0]).dmn_order
    data = {}
    for key_ in range(len(dmn_order)):
        roi = dmn_order[key_].split('_')[0]
        hemi = dmn_order[key_].split(' ')[1]
        data[f"{roi}_{hemi}"] = []
    
    first_flag = True
    for ix,id_ in enumerate(marmoset_ids):
        if first_flag:
            dstack_data=read_pkl(marmoset_ids[ix]).dmn_to_cortex
            first_flag=False
        else:
            dstack_data = np.dstack((dstack_data,read_pkl(marmoset_ids[ix]).dmn_to_cortex))
            
    if method == 'mean':
        return dstack_data.mean(2)
    elif method == 'dmn_order':
        return [i.split('_')[0]+'_'+i.split(' ')[-1] for i in dmn_order]
    else:         
        return dstack_data
    
sing_shell="singularity exec -B /mnt/WD10TB -B / /home/geoff/Desktop/containers/fmriprep_ciftify-1.3.2-2.3.3"
def save_conn_data(arr,fn,species='human',sing_shell=sing_shell,savedir="notebook_data/2_DMN_nodes/dmn_to_cortex"):
    
    if species == 'human':
        lh = "../data/myelinmap/S1200.lh.MyelinMap_BC_MSMAll.10k_fs_LR.s6.shape.gii"
        rh = "../data/myelinmap/S1200.rh.MyelinMap_BC_MSMAll.10k_fs_LR.s6.shape.gii"
    if species == 'marmoset':
        lh = "../data/myelinmap/surfFS.lh.myelinmap.10k.s2.shape.gii"
        rh = "../data/myelinmap/surfFS.rh.myelinmap.10k.s2.shape.gii"
    
    for myelinmap in [lh,rh]:
        if 'lh' in myelinmap:
            hemi = 'lh'
        else:
            hemi = 'rh'

        x = nib.load(myelinmap)
        if hemi == 'lh':
            y = arr[:10242]
        else:
            y = arr[10242:]
        x.add_gifti_data_array(nib.gifti.GiftiDataArray(y,datatype='NIFTI_TYPE_FLOAT32', intent='NIFTI_INTENT_POINTSET'))
        x.remove_gifti_data_array(0)

        nib.save(x,f"{savedir}/{fn}.{hemi}.shape.gii")
    
    print(f"Saving dscalar to: {savedir}/{fn}.dscalar.nii")
    os.system(f"{sing_shell} wb_command -cifti-create-dense-scalar {savedir}/{fn}.dscalar.nii -left-metric {savedir}/{fn}.lh.shape.gii -right-metric {savedir}/{fn}.rh.shape.gii")
    os.remove(f"{savedir}/{fn}.lh.shape.gii")
    os.remove(f"{savedir}/{fn}.rh.shape.gii")
```

Perform one sample t-test using different DMN-nodes for both species
 - this is repeated at various connectivity levels (for example)

```python
def randomise_DMN_nodes(species,idx1,outdir,greaterthr):
    if species == 'marmoset':
        ids_ = marmoset_ids
    else:
        ids_ = human_ids
        
    # Create input (-i) into randomise
    rsfc_data = get_rsfc(ids_,method="")
    data1 = rsfc_data[:,idx1,:]-greaterthr
    save_conn_data(data1,'onesample',species=species)
    
    savedir="notebook_data/2_DMN_nodes/dmn_to_cortex"
    cmd = f"{sing_shell} wb_command -cifti-convert -to-nifti {savedir}/onesample.dscalar.nii {savedir}/onesample.nii.gz"
    #print(cmd)
    os.system(cmd)
        
    # Get mask
    if species == 'marmoset':
        mask = '../data/data_marmoset/surfFS.atlasroi.10k.nii.gz'
    else:
        mask = '../data/data_human/cortex.nii.gz'    

    #print(f"mkdir {savedir}/{outdir}")
    os.system(f"mkdir {savedir}/{outdir}")
    cmd = f"{sing_shell} randomise -i {savedir}/onesample.nii.gz -m {mask} -1 -n 5000 -x --uncorrp -o {savedir}/{outdir}/randomise_results"
    #print(cmd)
    os.system(cmd)
    
    for i in os.listdir(f"{savedir}/{outdir}"):
        if 'nii.gz' in i:
            cmd = f"{sing_shell} wb_command -cifti-convert -from-nifti {savedir}/{outdir}/{i}  {savedir}/onesample.dscalar.nii {savedir}/{outdir}/{i.replace('nii.gz','dscalar.nii')} -reset-scalars"
            #print(cmd)
            os.system(cmd)
            
    os.remove(f"{savedir}/onesample.nii.gz")
    os.remove(f"{savedir}/onesample.dscalar.nii")


savedir="notebook_data/2_DMN_nodes/dmn_to_cortex"
onesamplettest_thresholds = {}
onesamplettest_thresholds['marmoset'] = np.arange(0,.3+.01,.01) # Probe onesample t-test [0<corr<.2; for marmoset data]
onesamplettest_thresholds['human'] = np.arange(0,.8+.01,.01)    # Probe onesample t-test [0<corr<.5; for human data]
# Loop through all permutation testing
for species in ['marmoset','human']:
    for thr_ in onesamplettest_thresholds[species]:
        thr_ = round(thr_,2)
        counter = 0
        for hemi in ['lh','rh']:
            for roi in ['dlPFC','PCC','PPC']:
                outdir_ = f"{species}_{hemi}-{roi}_onesamplettest_greaterthan{thr_}"
                if os.path.isdir(os.path.join(savedir,outdir_)):
                    continue
                else:
                    print(f"RUNNING: {species} randomization ({thr_}, {hemi}-{roi})"); 
                    randomise_DMN_nodes(species,counter,outdir_,thr_)
                counter += 1
```

```python
def get_onesamplettest_dscalars(species,hemi,thr):
    
    dscalar = {}
    for roi in ['dlPFC','PPC','PCC']:
        x1 = f"{species}_{hemi}-{roi}_onesamplettest_greaterthan0.2"
        x2 = f"connectivity.{species}.{roi}_{hemi}.thr-corrp-0.05.32k.dscalar.nii"
        dscalar[roi] = os.path.join(permutation_testing_dir,x1,x2)
    
    return dscalar

def dice_conn_coef(X,Y):
    """
    Take dscalar thresholded connectivity matrices and binarize it,
    and computes dice coefficient between X & Y
    """
    X = X>0
    Y = Y>0
    intersection = np.logical_and(X,Y)
    
    if (X.sum()+Y.sum()) == 0:
        return 0
    else:
        return 2.*intersection.sum() / (X.sum()+Y.sum())

def get_pairwise_dice_coef(species,hemi,thr):
    import itertools
    dscalars = get_onesamplettest_dscalars(species,hemi,thr)
    dice_coefs = {}
    for roi1_,roi2_ in itertools.combinations(dscalars.keys(),2):
        roi1 = f"notebook_data/2_DMN_nodes/dmn_to_cortex/{species}_{hemi}-{roi1_}_onesamplettest_greaterthan{thr}/connectivity.{species}.{roi1_}_{hemi}.thr-corrp-0.05.32k.dscalar.nii"
        roi2 = f"notebook_data/2_DMN_nodes/dmn_to_cortex/{species}_{hemi}-{roi2_}_onesamplettest_greaterthan{thr}/connectivity.{species}.{roi2_}_{hemi}.thr-corrp-0.05.32k.dscalar.nii"
        X = np.array(nib.load(roi1).get_fdata())
        Y = np.array(nib.load(roi2).get_fdata())
        dice_coefs[f"{roi1_}-{roi2_}"] = dice_conn_coef(X,Y)
        
    return dice_coefs
        

# thresholdlist for each species
# Save dice coefficients for each pair of ROIs, and for each species
thresholdlist = {}
dice_coefficients = {}
roipairs = ['dlPFC-PPC','dlPFC-PCC','PPC-PCC']
for hemi in ['lh','rh']:
    for key in onesamplettest.keys():
        thresholdlist[key] = []
        for pairs in roipairs:
            dice_coefficients[f"{key}_{hemi}_{pairs}"] = []
        for thr_ in onesamplettest_thresholds[key]:
            thr_ = round(thr_,2)
            thresholdlist[key].append(thr_)
            print(f"{key},{hemi},threshold-{thr_}")
            dice_coefs = get_pairwise_dice_coef(key,hemi,thr_)
            for pairs,coefs in dice_coefs.items():
                dice_coefficients[f"{key}_{hemi}_{pairs}"].append(coefs)
```
